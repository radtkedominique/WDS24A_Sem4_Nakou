### Fehlende Werte/NaN-Werte
Datensatz auf fehlende Werte analysieren. Dabei ergibt ein kurzer Blick in das Excel-Sheet, dass fehlende Werte nicht mit NaN gefüllt werden, sondern mit ".".
Wir ersetzen diese Werte durch np.nan und betrachten ihre Zusammenhänge.

#### Ergebnis
Bei 35 von 168 Produkten fehlen die Werte für das Jahr 2025.
Sonst treten keine NaN-Werte auf.


```python
destatis_df.info()
```

```python
(destatis_df == ".").sum()
```

```python
destatis_df = destatis_df.replace(".", np.nan)
```

```python
destatis_df.isnull().sum()
```

```python
missing_per_month = {}

for col in destatis_df.columns:
    missing_per_month[col] = set(destatis_df.loc[destatis_df[col].isna(), "Produkt"])
```

```python
missing_per_month["Januar.5"] == missing_per_month["Februar.5"] == missing_per_month["März.5"] == missing_per_month["April.5"] == missing_per_month["Mai.5"] == missing_per_month["Juni.5"]
```
