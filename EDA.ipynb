### Trendanalyse
Wir analysieren den Trend über alle Produkte hinweg, indem wir den Durchschnitt der Preisindizes im Januar 2024 mit dem Durchschnitt der Preisindizes im Juni 2025 vergleichen. Anschließend berechnen wir den prozentualen Trend für jedes Produkt und identifizieren die Top 10 Produkte mit dem stärksten Anstieg.


```python
destatis_df["Januar"].mean()
```

```python
destatis_df["Juni.5"].mean()
```

```python
trend_ueber_alle_produkte = destatis_df["Juni.5"].mean() / destatis_df["Januar"].mean()
print(trend_ueber_alle_produkte )
```

```python
trend_df = destatis_df[["Produkt", "Januar", "Juni.5"]].copy()
```

```python
trend_df[["Januar", "Juni.5"]] = trend_df[["Januar", "Juni.5"]].apply(pd.to_numeric, errors="coerce")
```

```python
trend_df["trend"] = ((trend_df["Juni.5"] - trend_df["Januar"]) / trend_df["Januar"]) * 100
```

```python
# ungültige Werte entfernen (NaN bei 2025)
trend_df = trend_df.replace([np.inf, -np.inf], np.nan).dropna(subset=["trend"])
```

```python
top10 = trend_df.sort_values("trend", ascending=False).head(10)
```

```python
top10
```

#### Produkte mit hoher Preisstabilität
Wir identifizieren die Produkte mit der geringsten Preisänderung, indem wir die Produkte mit den niedrigsten absoluten Trendwerten auswählen. Diese Produkte zeigen eine hohe Preisstabilität über den betrachteten Zeitraum.

```python
bottom10 = trend_df.sort_values("trend", ascending=True).head(10)
bottom10
```

```python
stabilste10 = (
    trend_df.assign(abs_trend=trend_df["trend"].abs())
            .sort_values("abs_trend", ascending=True)
            .head(10)
            .drop(columns="abs_trend")
)
stabilste10
```
