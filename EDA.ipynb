#### Dauerhafte Steigerung
Produkte, die eine dauerhafte Steigerung zeigen, haben einen positiven Trend über den gesamten Zeitraum.
Dafür verwenden wir das sheet "VÄR Vormonat_10-Steller", welches die monatlichen Preisänderungen (%) zeigt. Wir betrachten die Produkte, die in jedem Monat eine positive Preisänderung aufweisen, um die dauerhaft steigenden Produkte zu identifizieren.

```python
trend_vormonat_df = pd.read_excel('data_raw/sonderauswertung-nahrungsmittel.xlsx', sheet_name='VÄR Vormonat_10-Steller', skiprows=5)
```

### Vorbereitung der Daten
Bereinigung und Vorbereitung analog zu destatis_df, siehe oben für mehr Informationen.

```python
trend_vormonat_df.drop(index={168, 169, 170, 171}, inplace=True, errors='ignore')
trend_vormonat_df.rename(columns={'Unnamed: 0': 'COICOP-Index', 'Unnamed: 1': 'Produkt'}, inplace=True)
trend_vormonat_df = trend_vormonat_df.replace(".", np.nan)

# Monatsspalten identifizieren
month_columns = [col for col in trend_vormonat_df.columns if col not in ['COICOP-Index', 'Produkt']]

# Alle Spalten zu numeric konvertieren
for col in month_columns:
    trend_vormonat_df[col] = pd.to_numeric(trend_vormonat_df[col], errors='coerce')

print("Daten vorbereitet. Shape:", trend_vormonat_df.shape)
trend_vormonat_df.head()
```

### Identifikation: Produkte mit dauerhafter Steigerung
Ein Produkt zeigt "dauerhafte Steigerung", wenn alle monatlichen Preisänderungen positiv sind.
Da kein Produkt diese strikte Bedingung erfüllt, analysieren wir Produkte mit mindestens 80% positiven Monaten.

```python
# Produkte mit mindestens 80% positiven Monatswerten finden
dauerhafte_steigerungen = []

for idx, row in trend_vormonat_df.iterrows():
    produkt = row['Produkt']
    if pd.isna(produkt):
        continue

    # Alle Monatswerte für dieses Produkt
    werte = row[month_columns]

    # Zähle gültige und positive Werte
    valid_values = werte.dropna()
    positive_count = (werte > 0).sum()

    if len(valid_values) > 0:
        positive_ratio = positive_count / len(valid_values)

        # Kriterium: Mindestens 80% positive Monate
        if positive_ratio >= 0.8:
            dauerhafte_steigerungen.append({
                'Produkt': produkt,
                'Positive_Monate': positive_count,
                'Gesamt_Monate': len(valid_values),
                'Positive_Ratio_%': positive_ratio * 100,
                'Durchschnitts_Steigerung_%': valid_values.mean(),
                'Min_Steigerung_%': valid_values.min(),
                'Max_Steigerung_%': valid_values.max()
            })


dauerhafte_df = pd.DataFrame(dauerhafte_steigerungen).sort_values('Durchschnitts_Steigerung_%', ascending=False)

print(f"✓ Gefunden: {len(dauerhafte_df)} Produkte mit mindestens 80% positiven Monaten\n")
dauerhafte_df
```

### Alternative Kriterien für Steigerungsprodukte
Andere Perspektiven sind für die Identifikation von Produkten mit positiven Trends wichtig.
Wir betrachten die längste Sequenz von positiven Monaten, um Produkte zu identifizieren, die über einen längeren Zeitraum hinweg kontinuierlich gestiegen sind. Das Kriterium ist eine Sequenz von mindestens 12 aufeinanderfolgenden Monaten mit positiven Preisänderungen.

```python
# Kriterium: Längste Sequenz von positiven Monaten (mindestens 12 Monate hintereinander)
def find_longest_positive_sequence(values):
    """Finde die längste Sequenz von positiven Werten"""
    max_length = 0
    current_length = 0
    for val in values:
        if pd.notna(val) and val > 0:
            current_length += 1
            max_length = max(max_length, current_length)
        else:
            current_length = 0
    return max_length

longest_sequence_produkte = []
for idx, row in trend_vormonat_df.iterrows():
    produkt = row['Produkt']
    if pd.isna(produkt):
        continue

    werte = row[month_columns]
    longest = find_longest_positive_sequence(werte.values)

    if longest >= 12:
        valid_values = werte.dropna()
        longest_sequence_produkte.append({
            'Produkt': produkt,
            'Längste_positive_Sequenz': longest,
            'Durchschnitts_Steigerung_%': valid_values.mean() if len(valid_values) > 0 else 0,
            'Gesamt_Monate': len(valid_values)
        })

longest_seq_df = pd.DataFrame(longest_sequence_produkte).sort_values('Längste_positive_Sequenz', ascending=False)
print(f"\nGefunden: {len(longest_seq_df)} Produkte mit ≥12 Monaten positiver Sequenz\n")
if len(longest_seq_df) > 0:
    print(longest_seq_df.to_string(index=False))
else:
    print("Keine Produkte gefunden, die ≥12 Monate hintereinander positive Steigerungen aufzeigen.")
```

### Identifikation von Seitwärtsbewegungen
Neben dauerhaften Steigerungen sind auch Produkte mit Seitwärtsbewegungen interessant, da sie eine hohe Preisstabilität aufweisen. Statt mit starren Grenzwerten zu arbeiten, verwenden wir eine datengetriebene Kombination aus Kennzahlen und Ranking. So identifizieren wir die Produkte, die im Vergleich zum Gesamtdatensatz am ehesten eine Seitwärtsbewegung zeigen.

#### Beschreibung der Kennzahlen für Seitwärtsbewegungen

Zur Identifikation von Seitwärtsbewegungen wurden mehrere Kennzahlen berechnet.
Grundidee: Produkte mit **kleinen Werten** in diesen Kennzahlen zeigen über den Beobachtungszeitraum hinweg eine **hohe Preisstabilität** und damit eher eine Seitwärtsbewegung.

**Produkt**
Bezeichnung des betrachteten Lebensmittels.

**Seitwaerts_Score**
Gesamtscore zur Bewertung, wie stark ein Produkt eine Seitwärtsbewegung zeigt.
Der Score basiert auf den Rangplätzen mehrerer Kennzahlen im Gesamtdatensatz.
Ein **niedriger Score** bedeutet: Das Produkt gehört über mehrere Stabilitätskriterien hinweg zu den unauffälligen bzw. stabilen Produkten.
→ **Je kleiner der Score, desto eher Seitwärtsbewegung.**

**Nettoaenderung_%**
Absolute prozentuale Veränderung zwischen dem ersten und dem letzten verfügbaren Monatswert.
Damit wird gemessen, wie stark sich der Preisindex insgesamt über den gesamten Zeitraum verändert hat.
Ein kleiner Wert bedeutet, dass sich Anfangs- und Endwert nur wenig unterscheiden.
→ **Kleine Werte sprechen für Seitwärtsbewegung.**

**Robuste_Spannweite_%**
Relative Spannweite der Zeitreihe auf Basis des 10%- und 90%-Quantils, bezogen auf den Median.
Diese Kennzahl misst, wie stark die Werte im typischen Bereich streuen, ohne dass einzelne Ausreißer zu stark ins Gewicht fallen.
Ein kleiner Wert bedeutet, dass sich die Preisindizes über die Zeit nur in einem engen Bereich bewegen.
→ **Kleine Werte sprechen für eine stabile, seitwärts laufende Entwicklung.**

**Volatilitaet_MoM_%**
Standardabweichung der monatlichen relativen Veränderungen (*Month-over-Month*).
Sie beschreibt, wie stark die Monatsänderungen schwanken.
Ein kleiner Wert bedeutet, dass sich der Preisindex von Monat zu Monat nur gering verändert.
→ **Je niedriger die Volatilität, desto ruhiger verläuft die Zeitreihe.**

**Trendsteigung_rel_pro_Monat_%**
Relative Steigung einer linearen Trendgeraden pro Monat, bezogen auf den Median der Zeitreihe.
Diese Kennzahl zeigt, ob ein Produkt über die Zeit eher einen aufwärts- oder abwärtsgerichteten Trend hat.
Da für Seitwärtsbewegungen vor allem die Stärke des Trends relevant ist, wird hier der absolute Wert betrachtet.
Ein kleiner Wert bedeutet: Es gibt kaum einen klaren langfristigen Auf- oder Abwärtstrend.
→ **Kleine Werte sprechen für Seitwärtsbewegung.**

**Max_abs_MoM_%**
Größte absolute monatliche Veränderung innerhalb des Beobachtungszeitraums.
Diese Kennzahl zeigt, wie stark der Preisindex in seinem stärksten Einzelmonat ausgeschlagen hat.
Ein kleiner Wert bedeutet, dass es keine starken Sprünge zwischen zwei aufeinanderfolgenden Monaten gab.
→ **Kleine Werte sprechen für stabile Preisbewegungen.**

**Gueltige_Monate**
Anzahl der Monate, für die für das jeweilige Produkt tatsächlich Werte vorliegen.
Diese Kennzahl dient der Einordnung der Datenqualität: Je mehr gültige Monate vorhanden sind, desto belastbarer ist die Bewertung.

**Seitwaerts**
Boolesche Kennzeichnung, ob ein Produkt nach den gewählten Kriterien als Seitwärtsbewegung eingestuft wurde.
`True` bedeutet, dass das Produkt die definierten Stabilitätsanforderungen erfüllt.

#### Interpretation
Ein Produkt wird als besonders seitwärts laufend angesehen, wenn es:
- nur eine geringe Nettoveränderung zwischen Start- und Endwert zeigt,
- über die Zeit nur in einer engen Bandbreite schwankt,
- von Monat zu Monat nur geringe Veränderungen aufweist,
- keinen klaren Aufwärts- oder Abwärtstrend besitzt,
- und keine starken Einzelausschläge zeigt.

Damit beschreibt die Seitwärtsanalyse nicht nur "kaum Veränderung insgesamt", sondern auch eine **ruhige und trendarme Entwicklung über den gesamten Zeitraum**.

```python
# Datengrundlage: Indexniveau je Monat (destatis_df)
side_df = destatis_df.copy()

meta_cols = ["COICOP-Index", "Produkt"]
month_cols = [c for c in side_df.columns if c not in meta_cols]

# Sicherheitskonvertierung
side_df[month_cols] = side_df[month_cols].replace(".", np.nan)
side_df[month_cols] = side_df[month_cols].apply(pd.to_numeric, errors="coerce")

results = []

for _, row in side_df.iterrows():
    produkt = row["Produkt"]
    ts = pd.to_numeric(row[month_cols], errors="coerce")
    valid = ts.dropna().astype(float)

    # Datenqualitaet: genug Monate vorhanden
    n_valid = len(valid)
    miss_ratio = 1 - (n_valid / len(month_cols))
    if n_valid < 18 or miss_ratio > 0.20:
        continue

    start = valid.iloc[0]
    end = valid.iloc[-1]
    median = valid.median()

    if pd.isna(start) or pd.isna(end) or start == 0 or pd.isna(median) or median == 0:
        continue

    # 1) Nettoveraenderung
    net_change = abs(end / start - 1)

    # 2) Robuste Spannweite
    q10, q90 = valid.quantile([0.10, 0.90])
    robust_range = (q90 - q10) / median

    # 3) Volatilitaet der Monatsaenderungen
    mom = valid.pct_change().dropna()
    vol = mom.std() if len(mom) >= 2 else np.nan
    max_abs_mom = mom.abs().max() if len(mom) > 0 else np.nan

    # 4) Trendsteigung (linear) relativ zum Median
    if len(valid) >= 3:
        t = np.arange(len(valid))
        y = valid.to_numpy(dtype=float)
        slope = np.polyfit(t, y, 1)[0]
        slope_rel = abs(slope) / median
    else:
        slope_rel = np.nan

    results.append({
        "Produkt": produkt,
        "Nettoaenderung_%": net_change * 100,
        "Robuste_Spannweite_%": robust_range * 100,
        "Volatilitaet_MoM_%": vol * 100 if pd.notna(vol) else np.nan,
        "Trendsteigung_rel_pro_Monat_%": slope_rel * 100 if pd.notna(slope_rel) else np.nan,
        "Max_abs_MoM_%": max_abs_mom * 100 if pd.notna(max_abs_mom) else np.nan,
        "Gueltige_Monate": n_valid
    })

sideways_df = pd.DataFrame(results)

# Niedrige Werte in allen Kennzahlen sprechen fuer Seitwaertsbewegung
metric_cols = [
    "Nettoaenderung_%",
    "Robuste_Spannweite_%",
    "Volatilitaet_MoM_%",
    "Trendsteigung_rel_pro_Monat_%",
    "Max_abs_MoM_%"
]

for col in metric_cols:
    sideways_df[f"{col}_Rang"] = sideways_df[col].rank(pct=True, method="average")

sideways_df["Seitwaerts_Score"] = sideways_df[[f"{col}_Rang" for col in metric_cols]].mean(axis=1)

# Datengetriebene Guardrails: Produkte muessen in zentralen Stabilitaetskennzahlen
# unterhalb der jeweiligen Median-/40%-Schwellen liegen.
net_limit = sideways_df["Nettoaenderung_%"].quantile(0.40)
range_limit = sideways_df["Robuste_Spannweite_%"].quantile(0.40)
max_mom_limit = sideways_df["Max_abs_MoM_%"].quantile(0.50)
score_limit = sideways_df["Seitwaerts_Score"].quantile(0.20)

sideways_df["Seitwaerts"] = (
    (sideways_df["Nettoaenderung_%"] <= net_limit)
    & (sideways_df["Robuste_Spannweite_%"] <= range_limit)
    & (sideways_df["Max_abs_MoM_%"] <= max_mom_limit)
    & (sideways_df["Seitwaerts_Score"] <= score_limit)
)

# Falls die Guardrails immer noch zu streng sind, nehmen wir die 10 seitwaerts-aehnlichsten Produkte.
seitwaerts_produkte = sideways_df[sideways_df["Seitwaerts"]].copy()

if seitwaerts_produkte.empty:
    seitwaerts_produkte = sideways_df.nsmallest(10, "Seitwaerts_Score").copy()
    seitwaerts_produkte["Seitwaerts"] = True

seitwaerts_produkte = seitwaerts_produkte.sort_values("Seitwaerts_Score").reset_index(drop=True)

print(f"Gefundene Seitwaerts-Produkte: {len(seitwaerts_produkte)}")
print("\nKalibrierte Schwellenwerte:")
print(f"- Nettoaenderung <= {net_limit:.2f}%")
print(f"- Robuste Spannweite <= {range_limit:.2f}%")
print(f"- Max. absolute Monatsveraenderung <= {max_mom_limit:.2f}%")
print(f"- Seitwaerts-Score <= {score_limit:.3f}")

seitwaerts_produkte[[
    "Produkt",
    "Seitwaerts_Score",
    "Nettoaenderung_%",
    "Robuste_Spannweite_%",
    "Volatilitaet_MoM_%",
    "Trendsteigung_rel_pro_Monat_%",
    "Max_abs_MoM_%",
    "Gueltige_Monate"
]].head(20)
```

#### Trendwechsel Grafisch darstellen
Um die Trends einzelner Produkte besser zu verstehen, können wir die monatlichen Preisänderungen grafisch darstellen. Dafür erstellen wir eine Funktion, die die monatlichen VÄR-Vormonat-Werte für ein bestimmtes Produkt plottet. So können wir visuell erkennen, ob ein Produkt eine dauerhafte Steigerung, Seitwärtsbewegung oder starke Schwankungen aufweist.

```python
import matplotlib.pyplot as plt

def plot_produkt_trend(
    trend_vormonat_df: pd.DataFrame,
    produkt_name: str,
    start_year: int = None,
    figsize=(16, 5),
    marker="o"
):
    """
    Plottet die monatlichen VÄR-Vormonat-Werte für genau ein Produkt.

    Parameter:
    - trend_vormonat_df: DataFrame mit Spalten 'Produkt' und Monats-Spalten
    - produkt_name: exakter Produktname (oder Teilstring)
    - start_year: optional, z.B. 2021 (dann werden x-Labels wie 'Januar 2021' erzeugt)
    - figsize: Plot-Größe
    - marker: Marker-Stil
    """
    # Monatsspalten bestimmen (alles außer Metadaten)
    meta_cols = ["COICOP-Index", "Produkt"]
    month_cols = [c for c in trend_vormonat_df.columns if c not in meta_cols]

    # Produkt suchen: erst exakt, dann Teilstring
    exakter_treffer = trend_vormonat_df["Produkt"].astype(str).str.lower() == produkt_name.lower()
    subset = trend_vormonat_df[exakter_treffer]

    if subset.empty:
        teil_treffer = trend_vormonat_df["Produkt"].astype(str).str.contains(produkt_name, case=False, na=False)
        subset = trend_vormonat_df[teil_treffer]

    if subset.empty:
        raise ValueError(f"Kein Produkt gefunden für: '{produkt_name}'")

    if len(subset) > 1:
        print("Hinweis: Mehrere Treffer gefunden, es wird der erste verwendet.")
        print("Treffer-Beispiele:", subset["Produkt"].head(5).tolist())

    row = subset.iloc[0]
    produkt = row["Produkt"]

    # Werte numerisch
    y = pd.to_numeric(row[month_cols], errors="coerce")

    # Nur gültige Werte plotten
    valid_mask = y.notna()
    y_valid = y[valid_mask]
    cols_valid = [c for c, ok in zip(month_cols, valid_mask) if ok]

    # X-Labels erzeugen
    if start_year is not None:
        x_labels = []
        for i, c in enumerate(cols_valid):
            monat = str(c).split(".")[0]  # z.B. "Januar.2" -> "Januar"
            jahr = start_year + (i // 12)
            x_labels.append(f"{monat} {jahr}")
    else:
        x_labels = cols_valid

    # Plot
    plt.figure(figsize=figsize)
    plt.plot(x_labels, y_valid.values, marker=marker, linewidth=1.8)
    plt.title(f"Monatliche Preisänderung (VÄR Vormonat) - {produkt}")
    plt.xlabel("Monat")
    plt.ylabel("Veränderung ggü. Vormonat (%)")
    plt.grid(True, alpha=0.3)
    plt.xticks(rotation=60)
    plt.tight_layout()
    plt.show()
```

```python
plot_produkt_trend(trend_vormonat_df, "Butter")
plot_produkt_trend(trend_vormonat_df, "Vollmilch")
```
