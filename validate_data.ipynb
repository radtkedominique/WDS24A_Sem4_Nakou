```python
cpi_df = pd.read_csv('data_raw/cpi_de.csv')
```

```python
cpi_df
```

```python
cpi_df["Datum"] = pd.to_datetime(cpi_df["Datum"])
cpi_df = cpi_df.sort_values("Datum")

cpi_df["MoM"] = cpi_df["CPI"].pct_change() * 100  # in Prozent
print(cpi_df[["Datum", "CPI", "MoM"]])
```

```python

cpi_df
```

```python
cpi_df["Datum"] = pd.to_datetime(cpi_df["Datum"])
cpi_df = cpi_df.sort_values("Datum").copy()

cpi_df["YoY_%"] = (cpi_df["CPI"] / cpi_df["CPI"].shift(12) - 1) * 100
cpi_df["MoM_%"] = (cpi_df["CPI"] / cpi_df["CPI"].shift(1)  - 1) * 100

print(cpi_df[["Datum", "CPI", "MoM_%", "YoY_%"]].round(3))
```
